```python

def to_sentiment(rating):

    rating = int(rating)

    # Convert to class
    if rating <= 2:
        return 0
    elif rating == 3:
        return 1
    else:
        return 2

# Apply to the dataset
ds['sentiment'] = ds.score.apply(to_sentiment)
```

```python
from IPython.display import display
import ipywidgets as widgets
import pandas as pd

# Function to predict sentiment
def predict_sentiment(test_sentence):
    # Encode the test sentence
    predict_input = tokenizer.encode(test_sentence,
                                      truncation=True,
                                      padding=True,
                                      return_tensors="tf")

    # Predict sentiment
    tf_output = loaded_model.predict(predict_input)[0]
    tf_prediction = tf.nn.softmax(tf_output, axis=1)
    label = tf.argmax(tf_prediction, axis=1).numpy()[0]

    return labels[label]

# Define widgets
text_input = widgets.Text(placeholder='Enter a sentence')
button = widgets.Button(description='Predict')
output = widgets.Output()

# Define event handler for button click
def on_button_click(b):
    with output:
        output.clear_output()
        test_sentence = text_input.value
        prediction = predict_sentiment(test_sentence)
        data = {'Sentence': [test_sentence], 'Sentiment Prediction': [prediction]}
        df = pd.DataFrame(data)
        display(df)

# Attach event handler to button
button.on_click(on_button_click)

# Display widgets
display(text_input, button, output)
```

```python
from IPython.display import display
import ipywidgets as widgets

# Function to predict sentiment
def predict_sentiment(test_sentence):
    # Encode the test sentence
    predict_input = tokenizer.encode(test_sentence,
                                      truncation=True,
                                      padding=True,
                                      return_tensors="tf")

    # Predict sentiment
    tf_output = loaded_model.predict(predict_input)[0]
    tf_prediction = tf.nn.softmax(tf_output, axis=1)
    label = tf.argmax(tf_prediction, axis=1).numpy()[0]

    return labels[label]

# Define widgets
text_input = widgets.Text(placeholder='Enter a sentence')
button = widgets.Button(description='Predict')
output = widgets.Output()

# Define event handler for button click
def on_button_click(b):
    with output:
        output.clear_output()
        result = predict_sentiment(text_input.value)
        print(f'Sentiment prediction: {result}')

# Attach event handler to button
button.on_click(on_button_click)

# Display widgets
display(text_input, button, output)
```
